```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'Area':[2000,2500,2900,3100,3200],
    'Price':[350000,424000,433000,483000,497544]
}
df=pd.DataFrame(data)
df
```

Splitting the input and output data

```python
x= df[['Area']]
y=df['Price']
```

Import linear regression package

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_
print("Slope:",slope[0])
intercept=model.intercept_
print(f"Intercept:{intercept:.2f}")
print("Linear regression equation:")
print(f"y={slope[0]:.2f}*x+{intercept:.2f}")
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title("House price data")
```

```python
#predicted data
y_pred=model.predict(x)
plt.figure(figsize=(8,8))
plt.scatter(x,y)#actual data plot
plt.plot(x,y_pred)#linear regression line
plt.xlabel('Area')
plt.ylabel('Price')
plt.title("House price data")
plt.show()
```

```python
y_predict=model.predict([[3500]])
y_predict
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y,y_pred)
r2=r2_score(y,y_pred)
print(f"Mean Squared Error:{mse:.2f}")
print(f"R-squared:{r2:.2f}")
```

```python
data={
    "Years_of_experience":[5,7,8,11],
    "Education":['Bachelor','Bachelor','Masters','Masters'],
    "Salary":[35000,45000,55000,85000]
}
df=pd.DataFrame(data)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df
```

```python
x=df[['Years_of_experience','Education']]
y=df['Salary']
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_
intercept = model.intercept_
print("Slopes (coefficients):", slope)
print(f"Intercept: {intercept:.2f}")
print("\nLinear Regression Equation:")
print(f"Salary = {intercept:.2f} + "
      f"{slope[0]:.2f}*(Years_of_experience) + "
      f"{slope[1]:.2f}*(Education)")
```
