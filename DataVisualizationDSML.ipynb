```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['figure.dpi'] = 100

# For reproducibility
np.random.seed(42)

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv("/content/IRIS.csv")
```

```python
df.head(9)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal_length'],df['sepal_width'],alpha=0.6, color='steelblue', edgecolors='white', s=80)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Sepal Width', fontsize=12)
plt.title('Relationship between Sepal Length and Sepal Width', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
